```python
import pandas as pd 
import geopandas as gpd

from datetime import datetime as dt

import create_plots
```

```python
# read in data
ca = gpd.read_file("./example/county_level_shp/county_level.shp")
data = pd.read_csv("./example/weighted_averages.csv")

data["weighted_average"] = [float(i) for i in data["weighted_average"]]
data["county_fips"] = ['0' + str(i) for i in data["county_fips"]]
data['date_range_start'] = [dt.strptime(date,"%Y-%m-%dT%H:%M:%SZ").date() for date in data['date_range_start']]
```

```python
grouped = create_plots.group_data(data, "weighted_average", "date_range_start", "county_fips")
```

```python
heat_export, col_order = create_plots.export_heat_vals(
    grouped = grouped,
    cols = "weighted_average", # list of column names to compute
    date_col = "date_range_start", # We use the dates to for plot title
    group_col = "county_fips", # column name of data to join on
    map = ca,
    incl_dd = True,
    map_group_col = "fips" # column name of map to join on
)
```

```python
from functools import reduce

with open("fip_order.ts", 'w') as f:
    f.write("const fips = ['" + reduce(lambda acc, curr: f"{acc}','{curr}", col_order) + "'];")
```

```python
import json

with open('heat_data.ts', 'w') as f:
    f.write(json.dumps(heat_export))
```

```python
with open('heat_data.json', 'w') as f:
    f.write(heat_export.to_json())
```
